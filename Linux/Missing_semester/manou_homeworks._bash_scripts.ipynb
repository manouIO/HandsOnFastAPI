### **Exercise 1: Using the Pipe Command**


1. **Create a Bash script** named `process_files.sh` that performs the following tasks:
   - Lists all files in the current directory and pipes the output to `grep` to filter for files that contain the word "report" in their names.
   - Further, pipe this output to `wc -l` to count how many such files exist.
   
   Example command:
   ```bash
   ls | grep "report" | wc -l
   ```



2. **Modify your script** using `vim` to also print a message indicating how many report files were found.


3. **Edit your script again** to redirect the final output (the count of report files) into a file called `report_count.txt`.


### **Exercise 2: Using the Tee Command**




1. **Create a Bash script** named `log_commands.sh` that performs the following tasks:
   - Uses the `history` command to display your command history.
   - Pipes this output to `tee` to save it into a file called `command_history.txt`.
   
   Example command:
   ```bash
   history | tee command_history.txt
   ```

2. **Modify your script** using `vim` to include a message that indicates where the command history has been saved.

3. **Edit your script again** to append the last 10 commands from your history into another file called `last_commands.txt`.



### **Exercise 3: Using Find Command with Redirection**



1. **Create a Bash script** named `find_text_files.sh` that performs the following tasks:
   - Uses the `find` command to search for all `.txt` files in a specified directory (e.g., `/path/to/search`) and prints their names.
   
   Example command:
   ```bash
   find /path/to/search -name "*.txt"
   ```

2. **Modify your script** using `vim` to redirect this output into a file called `text_files_list.txt`.

3. **Edit your script again** to include a statement that counts how many `.txt` files were found and appends this count to `text_files_list.txt`.
