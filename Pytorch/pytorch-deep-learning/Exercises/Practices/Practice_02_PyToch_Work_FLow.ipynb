```python
# Create *known* parameters
weight=0.6
bias=0.3

#create
start=0
end=1
step=0.02
X=torch.arange(start,end,step).unsqueeze(dim=1) #input
y=weight*X+bias #ideal output, ideal because the parameters weight and bias are given

X[:10],y[:10],len(X),len(y)
```

### Splitting data into training and testing (One of the most important concept in ML in general)

 In the analogy of wrtting and exam,
 the `training set` (**The model learn from this data**) equivalent to course material(you have all the lessons, and examples), 

Then you have the `validation set` :**(the model get tuned on this data)** which is the mid term or practice exam, we ajust and we retry if the model does't perform well . 

In most cases, you might not use the validation set.

Then finally the `test set` **(the model gets evaluated on this data)** which is the final exam

**Generalization:** The ability of ML to behave well on Models or datas it hasn't see before


```python
#Let's create a training and testing set
train_split=int(0.8*len(X))
X_train,y_train=X[:train_split],y[:train_split] 
X_test,y_test=X[train_split:],y[train_split:]  

len(X_train),len(y_train),len(X_test),len(y_test)
```

### How might we better visualize our data?

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):
    '''
    Plot training data, test data and compares predictions.
    '''
    
    plt.figure(figsize=(10,7))
    # Plot training data in Blue

    plt.scatter(train_data,train_labels,c="b",s=4,label="Training Datas")

    #Plot testing data in Green
    plt.scatter(test_data,test_labels,c="g",s=4,label="Testing Datas")
    
    #Are they predictions?
    if predictions is not None:
       plt.scatter(test_data,predictions,c='r',s=4,label="predictions")

    #Legend
    plt.legend(prop={'size':14})

```

```python
plot_predictions()
```
